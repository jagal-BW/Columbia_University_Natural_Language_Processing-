```python
import urllib
```

```python
# Use urllib/requests package to read CNBC article through URL link
webdata = urllib.request.urlopen('https://www.cnbc.com/2019/01/17/netflix-price-hike-helps-disney-upcoming-streaming-service-analyst.html').read()
```

```python
# Use BeautifulSoup to extract text from the article.

from bs4 import BeautifulSoup
from bs4.element import Comment
def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True

def text_from_html(body):
    soup = BeautifulSoup(body, 'html.parser')
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)
    return u" ".join(t.strip() for t in visible_texts)

words=text_from_html(webdata)
print(words)
```

```python
# Use regular expression package to locate all matches of $ in the article

import re
print('$ amounts:')
print(re.findall('\$\d*\.?\d+?', words))

# the position of $ amounts in the article
first_pos=0
pos=[]
for i in range(words.count('$')):
    new_list = words[first_pos:]
    next_pos = new_list.index('$') + 1
    pos.append(first_pos + new_list.index('$'))
    first_pos += next_pos
print('positions: ')
print(pos)
```

```python
# Substitute numbers with # 
print(re.sub(r'[0-9]','#',words))
```

```python
# using regular expressions to count Netflix and Disney within text 
print('Netflix: '+str(len(re.findall('Netflix', words, flags=0))))
print('Disney: '+str(len(re.findall('Disney', words, flags=0))))
```

```python
import nltk
nltk.download('punkt')
```

```python
# Use NTLK to tokenize sentences and words

import re
from nltk import word_tokenize, sent_tokenize, ngrams, pos_tag, RegexpParser
from collections import Counter
```

```python
# tokenize sentences
sentences = sent_tokenize(words)
#print(sentences)
for sentence in sentences:
    print(sentence)
    print()
```

```python
#tokenize words from entire text (all text as imported) 
tokens_words=word_tokenize(words)
print(tokens_words)
```

```python
#tokenizing words within tokened sentences for observation 
for sentence in sentences:
    tokens = word_tokenize(sentence)
    print(tokens)
    print()
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
# Remove English stopwords (see output below)
stp_words = set(stopwords.words('english')) 
clean_words = [] 
```

```python
# Lemmatize and deduplicate unigrams into a dictionary (see output below)  
word_lemmatizer = WordNetLemmatizer()
deduplicate=[]
```

```python
# Use stemming for comparison to lemmatization (see output below)
#from nltk.stem.snowball import SnowballStemmer
#stemming = SnowballStemmer(language='english')
from nltk.stem.porter import PorterStemmer
stemming=PorterStemmer()
deduplicate2=[]
```

```python
#As you will see, Lemmatization is more precise than stemming --which is generally the case.   The two most commonly
#used stemming algorithms are PorterStemmer and SnowballStemmer.  I tried both and the outputs are similar.   
#Nontheless, both algorithms are both NOT as precise a lemmatization.   

for r in tokens_words: 
    if r not in stp_words: 
        clean_words.append(r) 
        if word_lemmatizer.lemmatize(r) not in deduplicate:
            deduplicate.append(word_lemmatizer.lemmatize(r))
        if stemming.stem(r) not in deduplicate2:
            deduplicate2.append(stemming.stem(r))
            
print('Tokens after removing all Stop Words: \n', clean_words,'\n')

print('Lemmatizing and Deduplicating unigrams into a dictionary: \n', deduplicate,'\n')
print('Length of the dictionary: ',len(deduplicate),'\n')
  

print('Stemming using PorterStemmer Algorithm: \n',deduplicate2,'\n')
print('Length of the dictionary: ',len(deduplicate2),'\n')
```

```python
# List and count n-grams for any given input n
def ListAndCount(tokens,n):
    for item in ngrams(tokens,n):
        print(item)
    print(Counter(ngrams(tokens,n)))

# n=4
ListAndCount(clean_words,4)
```
