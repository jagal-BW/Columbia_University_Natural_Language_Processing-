```python
import csv   #Creating a function to identify Duplicate Values.
seen = []    
with open('/Users/batman/Downloads/cb_sample.csv') as csv_infile:
  reader = csv.DictReader(csv_infile)
  for row in reader:
      if row['\ufeffname'] in seen:   
        print("Duplicate found; first occurrence is record", seen.index(row['\ufeffname'])+2, 
              "matches in record ", reader.line_num) 
      seen.append(row['\ufeffname'])   
```

```python
import csv   #creating a file with no the duplicate values 
seen=[]
with open('/Users/batman/Downloads/cb_sampleNew.csv', 'w') as csv_outfile: 
    fieldnames = ['\ufeffname', 'city', 'region', 'country','category_list', 'category_groups_list', 'cb_url', 'rank', 'domain']
    writer = csv.DictWriter(csv_outfile, fieldnames=fieldnames) 
    writer.writeheader()       
    with open('/Users/batman/Downloads/cb_sample.csv') as csv_infile:  
        reader = csv.DictReader(csv_infile)   
        for row in reader:   
            if not row['\ufeffname'] in seen:
                writer.writerow(row)
                seen.append(row['\ufeffname'])
```
